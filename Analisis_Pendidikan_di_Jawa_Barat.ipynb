# Mengumpulkan dan Menampilkan Data

## Mencari semua file dalam satu folder

Menggunakan library `glob` untuk mencari semua file dengan format `.csv` pada folder yang ditentukan.


```python
import glob

csv_files = glob.glob("/content/drive/MyDrive/Project/Data Science/Dataset/*.csv")
print(csv_files)
```

## Load dan Simpan Data

Buka semua file pada list file CSV yang telah dibuat: Disimpan dalam bentuk DataFrame dan digabungkan dalam dictionary


```python
import pandas as pd
import os

dataframes = {}

for file_path in csv_files:
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    df = pd.read_csv(file_path)
    dataframes[file_name] = df

for name, df in dataframes.items():
    print(f"DataFrame: {name}, Shape: {df.shape}")
```

## Memastikan semua file dibuka

Tampilkan key pada dictionary beserta banyaknya df yang dibuka


```python
print("Keys of the dataframes dictionary:")
print(dataframes.keys())
print("\nNumber of dataframes loaded:")
print(len(dataframes))
```

## Tampilkan data

Tampilkan semua dataframe yang dibuka


```python
for name, df in dataframes.items():
    print(f"Head of DataFrame: {name}")
    display(df.head())
    print("\n")
```
